```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataSA = pd.read_csv('C:/Pythondatasets/All_Valid_Datasets_Sarajevo_without_duplicates.csv',parse_dates=True, index_col='Date (LT)')
dataBL = pd.read_csv('C:/Pythondatasets/Banja Luka/All_Valid_Datasets_BanjaLuka_without_duplicates.csv',parse_dates=True, index_col='Date (LT)')
```

```python
plt.title("Maximum AQI value in Banja Luka from 2021 to 2023")
plt.xlabel("Year")
plt.ylabel("AQI")
dataBL['AQI'].plot()
plt.savefig('C:/Pythondatasets/test.png')
```

```python
plt.title("AQI value in Sarajevo from 2018 to 2023")
plt.xlabel("Year")
plt.ylabel("AQI")
dataSA['AQI'].plot(figsize=(12,5))
plt.savefig('C:/Pythondatasets/test_Sarajevo.png')
```

```python
AQIPoGodinama=dataSA.groupby('Year')['AQI'].max()
AQIPoGodinamaX=AQIPoGodinama.index.tolist()
AQIPoGodinamaY=AQIPoGodinama.values.tolist()
AQIPoGodinama
```

```python
def addlabels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], round(y[i], 1), ha='center')
```

```python
plt.title("Maximum AQI value in Sarajevo from 2018 to 2023")
plt.xlabel("Year")
plt.ylabel("AQI")
addlabels(AQIPoGodinamaX, AQIPoGodinamaY)
AQIPoGodinama.plot(kind="bar", color="red")
plt.savefig('C:/Pythondatasets/maks_vrijednost_AQI_po_godinama.png')
```

```python
januari=dataSA.loc[dataSA['Month']==1]
januariPoGodinama=januari.groupby('Year')['AQI'].max()
februari=dataSA.loc[dataSA['Month']==2]
februariPoGodinama=februari.groupby('Year')['AQI'].max()
decembri=dataSA.loc[dataSA['Month']==12]
decembriPoGodinama=decembri.groupby('Year')['AQI'].max()
```

```python
plt.title("Maximum AQI value in January in Sarajevo")
plt.xlabel("Year")
plt.ylabel("AQI")
x=januariPoGodinama.index.tolist()
y=januariPoGodinama.values.tolist()
addlabels(x, y)

januariPoGodinama.plot(kind="bar", color="red")
plt.savefig('C:/Pythondatasets/maks_vrijednost_AQI_januar_po_godinama.png')
```

```python
plt.title("Maximum AQI value in February in Sarajevo")
plt.xlabel("Year")
plt.ylabel("AQI")
x=februariPoGodinama.index.tolist()
y=februariPoGodinama.values.tolist()
addlabels(x, y)
februariPoGodinama.plot(kind="bar", color="red")
plt.savefig('C:/Pythondatasets/maks_vrijednost_AQI_februar_po_godinama.png')
```

```python
plt.title("Maximum AQI value in December in Sarajevo")
plt.xlabel("Year")
plt.ylabel("AQI")
x=decembriPoGodinama.index.tolist()
y=decembriPoGodinama.values.tolist()
addlabels(x, y)
decembriPoGodinama.plot(kind="bar", color="red")
plt.savefig('C:/Pythondatasets/maks_vrijednost_AQI_decembar_po_godinama.png')
```
