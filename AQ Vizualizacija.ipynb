```python
februari=data.loc[data['Month']==2]
februari
```

```python
februariPoGodinama=februari.groupby(['Year', 'date'])['AQI'].max()
februariPoGodinama
```

```python
januari=data.loc[data['Month']==1]
januari
```

```python
januariPoGodinama=januari.groupby(['Year', 'date'])['AQI'].max()
januariPoGodinama
```

```python
decembri=data.loc[data['Month']==12]
decembri
```

```python
decembriPoGodinama=decembri.groupby(['Year', 'date'])['AQI'].max()
decembriPoGodinama
```

```python
plt.title("Maksimalna AQI vrijednost tokom decembra, januara i februara od 2018. do 2023. godine")
plt.xlabel("Godina")
plt.ylabel("AQI")

ax = plt.gca()

januariPoGodinama.plot(kind="scatter", label="Januar", ax=ax)
februariPoGodinama.plot(label="Februar", ax=ax)
decembriPoGodinama.plot(label="Decembar", ax=ax)

plt.xticks(rotation=90)


plt.legend()
plt.xlabel("Datum")
```
