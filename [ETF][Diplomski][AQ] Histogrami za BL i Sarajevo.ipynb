```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataSA = pd.read_csv('C:/Pythondatasets/All_Valid_Datasets_Sarajevo_without_duplicates.csv')
dataBL = pd.read_csv('C:/Pythondatasets/Banja Luka/All_Valid_Datasets_BanjaLuka_without_duplicates.csv')
```

```python
import numpy as np
from scipy import stats

# Example AQI data
aqi_values = dataSA['AQI']

# Calculate the Z-Scores for AQI data
z_scores = np.abs(stats.zscore(aqi_values))

# Define a threshold for identifying outliers (e.g., Z-Score > 2)
threshold = 8

# Identify outliers
outliers = [aqi_values[i] for i in range(len(aqi_values)) if z_scores[i] > threshold]

# Print the identified outliers
print("Outliers:", outliers)
```

```python
import numpy as np

# Example AQI data
aqi_values = dataSA['AQI']

# Calculate the Median and Median Absolute Deviation (MAD)
median = np.median(aqi_values)
mad = np.median(np.abs(np.array(aqi_values) - median))

# Calculate the Modified Z-Scores for AQI data
modified_z_scores = np.abs(0.6745 * (np.array(aqi_values) - median) / mad)

# Define a threshold for identifying outliers (e.g., Modified Z-Score > 3)
threshold = 8

# Identify outliers
outliers = [aqi_values[i] for i in range(len(aqi_values)) if modified_z_scores[i] > threshold]

# Print the identified outliers
print("Outliers:", outliers)
```
