```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data_sa = pd.read_csv('C:/Pythondatasets/All_Valid_Datasets_Sarajevo_sa_datumom - Copy.csv')
data_sa.columns
```

```python
data_bl = pd.read_csv('C:/Pythondatasets/Banja Luka/All_Valid_Datasets_Banja_Luka_sa_datumom - Copy.csv')
data_bl
```

```python
frames = [data_sa, data_bl]  
  
data = pd.concat(frames)  
data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
X=data[['Year', 'Month', 'Hour']]
Y=data['AQI Category']

X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.33, random_state=70)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)

# Logistic Regression

log_reg=LogisticRegression().fit(X_train, Y_train)

train_pred=log_reg.predict(X_train)
print("Model accuracy on train is: ", accuracy_score(Y_train, train_pred))

test_pred=log_reg.predict(X_test)
print("Model accuracy on test is: ", accuracy_score(Y_test, test_pred))
print("-"*50)

print("KappaScore is ", metrics.cohen_kappa_score(Y_test, test_pred))
```
