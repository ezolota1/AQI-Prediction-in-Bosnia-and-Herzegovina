```python
def discrete_time(date_now, date_before):

    delta=(date_now-date_before).total_seconds()/3600
    return delta-1
```

```python
date1=datetime.datetime(2018,5,22,18,0,0)
date2=datetime.datetime(2018,5,22,17,0,0)
print(discrete_time(date1, date2, 1))
```
