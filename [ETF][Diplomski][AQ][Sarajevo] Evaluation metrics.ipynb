```python
import matplotlib.pyplot as plt
import numpy as np

# Metric values for each model
models = ['Linear Regression', 'ARIMA', 'SARIMA']
rmse_values = [10.005, 9.883, 13.147]
mse_values = [100.105, 6.345, 8.849]
mae_values = [6.334, 97.667, 172.854]
r_square_values = [0.973, 0.969, 0.943]

# Number of models
num_models = len(models)

# Set the width of each bar
bar_width = 0.2

# Define the x-axis positions for the bars
x = np.arange(num_models)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Plot RMSE values
ax.bar(x - bar_width, rmse_values, width=bar_width, label='RMSE', color='skyblue')

# Plot MSE values
ax.bar(x, mse_values, width=bar_width, label='MSE', color='lightcoral')

# Plot MAE values
ax.bar(x + bar_width, mae_values, width=bar_width, label='MAE', color='lightgreen')

# Plot R-Square values
ax.bar(x + 2 * bar_width, r_square_values, width=bar_width, label='R-Square', color='lightseagreen')

# Set x-axis labels and title
ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(models, rotation=15)
ax.set_xlabel('Models')
ax.set_title('Evaluation Metrics for Air Quality Prediction in Sarajevo')

# Add a common y-axis label
ax.set_ylabel('Metric Value')

# Add a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.savefig('C:/Pythondatasets/Evaluation_Metrics_Prediction_Sarajevo.png')
plt.show()
```
