```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('C:/Pythondatasets/All_Valid_Datasets_Sarajevo_sa_datumom - Copy.csv')
data
```

```python
januari = data.loc[(data["Month"] == 1)]
januari
```

```python
# najveci AQI u januaru tokom godina

najveciAQIJanuar=januari[januari.AQI == januari.AQI.max()]
najveciAQIJanuar
```

```python
najgoriDanUJanuaru=januari.loc[januari['date']==najveciAQIJanuar['date'].item()]

najgoriDanUJanuaru
```

```python
najgoriDanUJanuaru.hist()
```

```python
najgoriDanUJanuaruPoSatima=najgoriDanUJanuaru.groupby('Hour')['AQI'].max()
najgoriDanUJanuaruPoSatima
```

```python
def addlabels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], round(y[i], 1), ha='center')
```

```python
plt.title("AQI vrijednost za najgori dan u januaru po satima")
plt.xlabel("Sat")
plt.ylabel("AQI")
x=najgoriDanUJanuaruPoSatima.index.tolist()
y=najgoriDanUJanuaruPoSatima.values.tolist()
addlabels(x, y)
najgoriDanUJanuaruPoSatima.plot(kind="bar", color="red")
```

```python
februari = data.loc[(data["Month"] == 2)]
najveciAQIFebruar=februari[februari.AQI == februari.AQI.max()]
najgoriDanUFebruaru=februari.loc[februari['date']==najveciAQIFebruar['date'].item()]
najgoriDanUFebruaruPoSatima=najgoriDanUFebruaru.groupby('Hour')['AQI'].max()

plt.title("AQI vrijednost za najgori dan u februaru po satima")
plt.xlabel("Sat")
plt.ylabel("AQI")
x=najgoriDanUFebruaruPoSatima.index.tolist()
y=najgoriDanUFebruaruPoSatima.values.tolist()
addlabels(x, y)
najgoriDanUFebruaruPoSatima.plot(kind="bar", color="red")
```
