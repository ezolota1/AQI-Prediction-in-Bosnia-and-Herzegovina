```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('C:/Pythondatasets/All_Valid_Datasets_Sarajevo_sa_datumom - Copy.csv')
data
```

```python
from sklearn.utils import shuffle
data=shuffle(data, random_state=42)
```

```python
div=int(data.shape[0]/4)
```

```python
train=data.iloc[:3*div+1, :]
test=data.iloc[3*div+1:]
train
```

```python
train['AQI'].mean()
```

```python
test['Simple_Mean']=train['AQI'].mean()
test
```

```python
from sklearn.metrics import mean_absolute_error as MAE
```

```python
simple_mean_error=MAE(test['AQI'], test['Simple_Mean'])
simple_mean_error
```

```python
# PROSJEČAN AQI PO MJESECU
out_type=pd.pivot_table(train, values='AQI', index='Month', aggfunc=np.mean)
out_type
```

```python
test['Out_Month_Mean']=0
for i in train['Month'].unique():
    test['Out_Month_Mean'][test['Month']==i]=train['AQI'][train['Month']==i].mean()
```

```python
test
```

```python
out_type_error=MAE(test['AQI'], test['Out_Month_Mean'])
out_type_error
```

```python
# PROSJEČAN AQI PO GODINI
out_year=pd.pivot_table(train, values='AQI', index='Year', aggfunc=np.mean)

test['Out_Year_Mean']=0
for i in train['Year'].unique():
    test['Out_Year_Mean'][test['Year']==i]=train['AQI'][train['Year']==i].mean()

out_type_error=MAE(test['AQI'], test['Out_Year_Mean'])
out_type_error
```

```python
# PREDIKCIJA PO VIŠE KOLONA (GODINA I MJESEC)

combo=pd.pivot_table(train, values='AQI', index=['Year', 'Month'], aggfunc=np.mean)

test['Super_Mean']=0
s2='Year'
s1='Month'
for i in test[s1].unique():
    for j in test[s2].unique():
        test['Super_Mean'][(test[s1]==i) & (test[s2]==j)]=train['AQI'][(train[s1]==i) & (train[s2]==j)].mean()

out_type_error=MAE(test['AQI'], test['Super_Mean'])
out_type_error
```

```python
# PROSJEČAN AQI PO SATU
out_hour=pd.pivot_table(train, values='AQI', index='Hour', aggfunc=np.mean)

test['Out_Hour_Mean']=0
for i in train['Hour'].unique():
    test['Out_Hour_Mean'][test['Hour']==i]=train['AQI'][train['Hour']==i].mean()

out_type_error=MAE(test['AQI'], test['Out_Hour_Mean'])
out_type_error
```

```python
# PREDIKCIJA PO VIŠE KOLONA (MJESEC I SAT)

combo=pd.pivot_table(train, values='AQI', index=['Month', 'Hour'], aggfunc=np.mean)

test['Super_Mean_Hour']=0
s2='Month'
s1='Hour'
for i in test[s1].unique():
    for j in test[s2].unique():
        test['Super_Mean_Hour'][(test[s1]==i) & (test[s2]==j)]=train['AQI'][(train[s1]==i) & (train[s2]==j)].mean()

out_type_error=MAE(test['AQI'], test['Super_Mean_Hour'])
out_type_error
```

```python
# PREDIKCIJA PO VIŠE KOLONA (GODINA, MJESEC I SAT)

combo=pd.pivot_table(train, values='AQI', index=['Year', 'Month', 'Hour'], aggfunc=np.mean)

test['Triple_Mean']=0
s2='Year'
s1='Month'
s0='Hour'
for z in test[s0].unique():
    for i in test[s1].unique():
        for j in test[s2].unique():
            test['Triple_Mean'][(test[s0]==z) & (test[s1]==i) & (test[s2]==j)]=train['AQI'][(train[s0]==z) & (train[s1]==i) & (train[s2]==j)].mean()
            
out_type_error=MAE(test['AQI'], test['Triple_Mean'])
out_type_error
```
