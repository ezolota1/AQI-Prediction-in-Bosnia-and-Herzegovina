```python
plt.figure(figsize=(10, 6))
plt.scatter(dataSA['AQI'], dataSA['Raw Conc.'], marker='o', color='blue', alpha=0.7)
plt.xlabel("AQI")
plt.ylabel('PM2.5')
plt.title("Scatter plot of AQI and PM2.5 in Sarajevo from 2018 to 2023")
plt.grid(True)
plt.tight_layout()
plt.savefig('C:/Pythondatasets/scatter_plot_AQI.png')
plt.show()
```

```python
import numpy as np
from scipy import stats

# Example AQI data
aqi_values = dataSA['AQI']

# Calculate the Z-Scores for AQI data
z_scores = np.abs(stats.zscore(aqi_values))

# Define a threshold for identifying outliers (e.g., Z-Score > 2)
threshold = 2

# Identify outliers
outliers = [aqi_values[i] for i in range(len(aqi_values)) if z_scores[i] > threshold]

# Print the identified outliers
print("Outliers:", outliers)
```
