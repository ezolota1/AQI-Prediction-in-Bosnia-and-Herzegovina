```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import timedelta64
import datetime
```

```python
data = pd.read_csv('C:/Pythondatasets/Banja Luka/All_Valid_Datasets_BanjaLuka_without_duplicates.csv', parse_dates=True)
data
```

```python
data = data.loc[(data["AQI"] > 0)]
data
```

```python
def discrete_time(date_now, date_before):
    
    delta=(date_now-date_before).total_seconds()/3600
    return delta-1
```

```python
data['pomocna'] = data['Date (LT)'].shift()
data
```

```python
data.insert(loc=15, column="diskretno_vrijeme", value=range(0, 0 + len(data)))
data
```

```python
for i in data.index:
    data['diskretno_vrijeme'][i]=data['diskretno_vrijeme'][i]+discrete_time(data['Date (LT)'][i], data['pomocna'][i])

data
```

```python
data['pomocna'] = pd.to_datetime(data['pomocna'], errors='coerce')
```

```python
data.to_csv('C:/Pythondatasets/Banja Luka/Dataset_BanjaLuka_Sa_Diskretnim_Vremenom.csv', index=False)
```
