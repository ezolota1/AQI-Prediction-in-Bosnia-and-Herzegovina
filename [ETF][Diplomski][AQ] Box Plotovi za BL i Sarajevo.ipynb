```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('C:/Pythondatasets/All_Valid_Datasets_Sarajevo_without_duplicates.csv')
data
```

```python
data = pd.read_csv('C:/Pythondatasets/Banja Luka/All_Valid_Datasets_BanjaLuka_without_duplicates.csv')
data
```

```python
plt.boxplot(data['AQI'])  
plt.show()  
```

```python
df_out2018 = data.loc[data['Year']==2018]
df_out2019 = data.loc[data['Year']==2019]
df_out2020 = data.loc[data['Year']==2020]
df_out2021 = data.loc[data['Year']==2021]
df_out2022 = data.loc[data['Year']==2022]
df_out2023 = data.loc[data['Year']==2023]
```

```python
AQI2018=df_out2018['AQI']
AQI2019=df_out2019['AQI']
AQI2020=df_out2020['AQI']
AQI2021=df_out2021['AQI']
AQI2022=df_out2022['AQI']
AQI2023=df_out2023['AQI']
```

```python
plt.title("Box plot of AQI in Sarajevo from 2018 to 2023")
plt.xlabel("Year")
plt.ylabel("AQI")
plt.boxplot([AQI2018, AQI2019, AQI2020, AQI2021, AQI2022, AQI2023], labels=['2018', '2019', '2020', '2021', '2022', '2023']) 
plt.savefig('C:/Pythondatasets/box_plot_AQI.png', dpi = 150, bbox_inches='tight')
plt.show()  
```

```python
plt.title("Box plot of AQI in Banja Luka from 2021 to 2023")
plt.xlabel("Year")
plt.ylabel("AQI")
plt.boxplot([AQI2021, AQI2022, AQI2023], labels=['2021', '2022', '2023']) 
plt.savefig('C:/Pythondatasets/box_plot_AQI_BanjaLuka.png', dpi = 150, bbox_inches='tight')
plt.show()  
```

```python
PM_25_2018=df_out2018['Raw Conc.']
PM_25_2019=df_out2019['Raw Conc.']
PM_25_2020=df_out2020['Raw Conc.']
PM_25_2021=df_out2021['Raw Conc.']
PM_25_2022=df_out2022['Raw Conc.']
PM_25_2023=df_out2023['Raw Conc.']
```

```python
plt.title("Box plot of PM2.5 in Sarajevo from 2018 to 2023")
plt.xlabel("Year")
plt.ylabel("PM2.5")
plt.boxplot([PM_25_2018, PM_25_2019, PM_25_2020, PM_25_2021, PM_25_2022, PM_25_2023], labels=['2018', '2019', '2020', '2021', '2022', '2023']) 
plt.savefig('C:/Pythondatasets/box_plot_PM25.png', dpi = 150, bbox_inches='tight')
plt.show()
```

```python
plt.title("Box plot of PM2.5 in Banja Luka from 2021 to 2023")
plt.xlabel("Year")
plt.ylabel("PM2.5")
plt.boxplot([PM_25_2021, PM_25_2022, PM_25_2023], labels=['2021', '2022', '2023']) 
plt.savefig('C:/Pythondatasets/box_plot_PM25_BanjaLuka.png', dpi = 150, bbox_inches='tight')
plt.show()
```
